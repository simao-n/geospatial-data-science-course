# Introduction to OpenStreetMap (OSM)
OpenStreetMap (OSM) is a collaborative, open-source mapping platform that provides structured geospatial data from around the world. Unlike traditional maps, OSM data can be downloaded, analyzed, and improved, making it a critical resource for urban mobility projects. This notebook introduces OSM and shows how to access its data using Python and the OSMnx library.

````python
# Install OSMnx required library
!pip install osmnx
````

````python
import osmnx as ox
G = ox.graph_from_place("Lisbon, Portugal", network_type="drive")
ox.plot_graph(G)

# note: This code can be slow.
````

You can also load different types of networks depending on your needs:

```python
G = ox.graph_from_place('Lisbon, Portugal', network_type='bike')
ox.plot_graph(G)
```

Available types: `'walk'`, `'bike'`, `'drive'`, `'drive_service'`, `'all'`, `'all_private'`.

````python
# You can also remove the nodes to have a clean visualization
ox.plot_graph(
    G,
    node_size=0,
    node_color='none'
)
````
